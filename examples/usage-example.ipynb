```python
import os

import pandas as pd

from visicom import Visicom
```

```python
api = Visicom(token=os.environ["VISICOM_KEY"])
```

```python
df = pd.read_excel("../assets/data.xlsx")
```

```python
df
```

```python
df["address"] = (
    df["Oblast"].fillna("") + " область, "
    + df["Rayon"].fillna("") + " район, "
    + df["type_Settlement"].fillna("") + " " 
    + df["Settlement"].fillna("") + ", " 
    + df["Street"].fillna("") 
)
```

```python
df["address"] = (
    df["address"]
    .str.replace("\s+", " ", regex=True)
    .str.replace(", район,", ",")
    .str.strip(", ")
)
```

```python
df
```

```python
df["coordinates"] = df["address"].apply(api.geocode)
```

```python
df["x"] = df["coordinates"].str[0]
df["y"] = df["coordinates"].str[1]
```

```python
df.loc[:, ["IDF", "address", "x", "y"]]
```

```python
df.loc[:, ["IDF", "address", "x", "y"]].to_csv("../assets/geocoded.csv", index=False)
```
